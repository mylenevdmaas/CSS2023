```python
from pyvis.network import Network
import networkx as nx
import pandas as pd
import random
```

```python
dti_matrix = pd.read_csv('DTI_avg_log.csv', header = None)
```

```python
dti_matrix
```

```python
dti_matrix_avg = (dti_matrix + dti_matrix.T) / 2
```

```python
dti_matrix_avg
```

```python
def create_nodes(n):
    values = []
    color = []
    for i in range(n):
        value = random.choice([1, -1])
        values.append(value)
        if value == 1:
            color.append('#FFFF00')
        else:
            color.append('#A020F0')
    return color
```

```python
def positioning(matrix, n):
    links = np.zeros(n)
    volume = np.zeros(n)
    iteration_list = np.arange(0, n ,1)
    x_list = np.zeros(n)
    y_list = np.zeros(n)
    for i in iteration_list:
        for j in iteration_list:
            if matrix[i][j] > 0:
                volume[i] += 1
    for i in iteration_list:
        links[i] = 89 - volume[i]
        judge = random.choice([1, -1])
        judge1 = random.choice([1, -1])
        destination_random = random.uniform(0.8,1.0)
        destination_random1 = random.uniform(0.8,1.0)
        x_list[i] = 400 + links[i] * judge * destination_random
        y_list[i] = 400 + links[i] * judge1 * destination_random1
        volume[i] = int(volume[i]**2)
    volume = volume.tolist()
    x_list = x_list.tolist()
    y_list = y_list.tolist()
    return volume, x_list, y_list
```

```python
network = Network(height = 600, width = 600, notebook = True, bgcolor="#222222", font_color="white")
node_list = np.arange(1 , 91, 1)
color = create_nodes(90)
network.toggle_hide_edges_on_drag(False)
network.barnes_hut()
volume, x_list, y_list = positioning(dti_matrix_avg, 90)
network.add_nodes(node_list, size = volume
                           , x = x_list
                           , y = y_list
                           , color = create_nodes(90))
iteration_list = np.arange(0, 90, 1)
edge_data = []
for i in range(0, 90):
    for j in range(0, 90):
        if (i < j) and dti_matrix_avg[i][j]>0:
            edge_data.append((i+1, j+1, dti_matrix_avg[i][j]))
                
for e in edge_data:
    network.add_edge(e[0], e[1])
        
        

# Add multiple edges using a list comprehension
#[network.add_edge(source, target) for source, target in edge_data]

neighbor_map = network.get_adj_list()

# add neighbor data to node hover data
for node in network.nodes:
    node["value"] = len(neighbor_map[node["id"]])

network.show("Network_of_brain.html")
```
